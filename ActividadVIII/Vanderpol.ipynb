```python
def vectorfield(w, t, p):
    """
   
   Vector Vanderpol
   
    Arguments:
        w :  vector of the state variables:
                  w = [x,y]
        t :  time
        p :  vector of the parameters:
                  p = [u]
    """
    X, Y = w

    # Create f = (x',y'):
    f = [Y,
         p*(1.0 - (X**2.0))*Y -X]
    return f
```

```python
# Use ODEINT to solve the differential equations defined by the vector field
from scipy.integrate import odeint
import numpy as np


#u
u = 4

# Initial conditions

X = 2
Y =  -3

stoptime = 1000
numpoints = 500

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
w0 = [X, Y]

# Call the ODE solver.
wsol = odeint(vectorfield, w0, t, args=(u,))

with open('Vanderpol1.dat', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```

```python
import numpy as np
def forced(w, t, p):
    """
    Defines the differential equation for the Van der Pol oscillation system.

    Arguments:
        w :  vector of the state variables:
                  w = [x,y]
        t :  time
        p :  vector of the parameters:
                  p = [mu]
    """
    x, y = w
    
    A, m, u = p

    # Create f = (x',y'):
    f = [y,
         u*(1.0 - x**2.0) * y - x + A*np.sin(m*t)]
    return f
```

```python
 #Use ODEINT to solve the differential equations defined by the vector field
from scipy.integrate import odeint
from numpy import *

# Parameter values

u = 8.53
A = 1.2
m = 2.0*pi/10.0

# Initial conditions

x = 1.0
y = -10.0

stoptime = 1000.0
numpoints = 2500

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
w0 = [x, y]

p=[A,m,u]

# Call the ODE solver.
wsol = odeint(forced, w0, t, args=(p,))

with open('VL.dat', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```
