```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('sargento_201117.csv', skiprows=2, sep=',')
df2 = pd.read_csv('sargento-salinidad-201117.csv', skiprows=3, sep=',')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.columns= ['Num', 'Fecha', 'Abs Pres, kPa', 'Temp', 'NAgua']
```

```python
df1.head(2400)
```

```python
df2.columns= ['Num', 'Fecha', 'Cond High Rng,Scm', 'Temp', 'Specific Conductance,Scm', 'Salinity']
```

```python
df2.head()
```

```python
df1=df1.drop(df1.index[[2393]])
```

```python
df1['Fecha'] = pd.to_datetime(df1['Fecha'], format='%m/%d/%Y %H:%M:%S')
df1['month'] = df1['Fecha'].dt.month
df1.head()
```

```python
df2['Fecha'] = pd.to_datetime(df2['Fecha'], format='%m/%d/%Y %H:%M:%S')
df2['month'] = df2['Fecha'].dt.month
df2.head()
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3=pd.concat([df1['Fecha'], df2['Salinity'], df1['month'], df2['Temp'], df1['NAgua']], axis =1)
```

```python
df3.head()
```

```python
ax = sns.boxplot(x="month", y="NAgua", data=df3)
plt.show()
```

```python
ax = sns.boxplot(x="month", y="Salinity", data=df3)
plt.show()
```

```python
ax = sns.boxplot(x="month", y="Temp", data=df3)
plt.show()
```

```python
sns.set(style="darkgrid", color_codes=True)

tips = sns.load_dataset("tips")
g = sns.jointplot("Salinity", "NAgua", data=df3, kind="reg",
                  color="r", size=7)

plt.show(g)
```

```python
fig, ax1 = plt.subplots()
t = df1['Num']
s1 = df3['NAgua']
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time (s)')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('metros', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
s2 = df3['Salinity']
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('ppt', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
```
